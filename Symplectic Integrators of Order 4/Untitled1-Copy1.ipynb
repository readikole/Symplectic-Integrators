```python
x1_vals = []
x2_vals = []
```

```python
# seed random number generator
seed(1)
# generate random numbers between 0-1
for _ in range(1000):
    value = random()
    x1_vals.append(value)
```

```python
# seed random number generator
seed(1)
# generate random numbers between 0-1
for _ in range(1000):
    value = random()
    x2_vals.append(value)
```

```python
x1_vals = np.array(x1_vals)
x2_vals = np.array(x2_vals)
```

```python
i = 0
for idx in range(len(x1_vals)):
#intial conditions 
    x1[0] = x1_vals[idx]
    x2[0] = x2_vals[idx]
    K =0.8
    for i in range(N-1):
        x1[i+1] = (x1[i] + x2[i] + (K/(2*np.pi))*np.sin(2*np.pi*x1[i]))%1
        x2[i+1] = (x2[i] + (K/(2*np.pi))*np.sin(2*np.pi*x1[i]))%1
    plt.scatter(x1, x2, s=0.0001)
    plt.title(r'\textbf{The phase portrait of the $2-D$ standard map}')
    plt.xlabel(r'$x_{1}$')
    plt.ylabel(r'$x_{2}$')
    #plt.xlim(0.3, 0.7)
    #plt.ylim(0.4, 0.65)
```
